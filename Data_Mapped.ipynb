###Data Visualization

```python
import numpy as np
import pandas as ps
import plotly as pt
import matplotlib.pyplot as plt
```

```python
df = ps.read_csv("/content/2014HuriData.csv")

df
```

```python
# prompt: plot the paths using scatter mapbox with mode markers+lines based on the long and lat positions given in df distinguish path by color. Link points with the same name through lines

import plotly.express as px

fig = px.scatter_mapbox(df, lat="lat", lon="long", color="name",
                        hover_name="name", zoom=3, height=600,
                        mapbox_style="carto-positron",
                        center={"lat": df.lat.mean(), "lon": df.long.mean()})

fig.update_traces(mode="markers+lines")

fig.show()
```
