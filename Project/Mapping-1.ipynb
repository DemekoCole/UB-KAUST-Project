```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("HurriData.csv")

df.dropna(
    axis=0,
    how='any',
    #thresh=None,
    #subset=None,
    inplace=True
)

color_scale = [(0, 'orange'), (1,'red')]

fig = px.scatter_mapbox(df, 
                        lat="lat", 
                        lon="long", 
                        hover_name="name", 
                        hover_data=["name", "category","year"],
                        color="name",
                        color_continuous_scale=color_scale,
                        
                        #mode="marker+lines",
                        zoom=8, 
                        height=800,
                        width=800)
fig.update_layout(mapbox_style="satellite")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
import plotly.graph_objects as go

df = pd.read_csv("HurriData.csv")

lond=df.long
latd=df.lat
fig = go.Figure(go.Scattermapbox(
    mode = "markers+lines",
    lon = lond,
    lat = latd,
    marker = {'color':"name",'size': 10}))
fig.add_trace(go.Scattermapbox(
    mode = "markers+lines",
    lon = lond,
    lat = latd,
    marker = {'color':"name",'size': 10}))

fig.update_layout(
    margin ={'l':0,'t':0,'b':0,'r':0},
    mapbox = {
        'center': {'lon': 10, 'lat': 10},
        'style': "open-street-map",
        'center': {'lon': -20, 'lat': -20},
        'zoom': 1})

fig.show()
```
